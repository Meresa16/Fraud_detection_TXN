```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/processed/bank_transactions_features.csv")

# Transaction count per customer in last 24h (simplified)
df['txn_count_24h'] = df.groupby('customer_id')['txn_hour'].transform(lambda x: x.rolling(24, min_periods=1).count())
df['amount_dev'] = df['amount'] - df['customer_avg_amount']
df['new_device'] = df.groupby('customer_id')['device_id'].transform(lambda x: x != x.shift(1)).astype(int)
df['new_location'] = df.groupby('customer_id')['location'].transform(lambda x: x != x.shift(1)).astype(int)

df.to_csv("../data/processed/bank_transactions_advanced_features.csv", index=False)
```
